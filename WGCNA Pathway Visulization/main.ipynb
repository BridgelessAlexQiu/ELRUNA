```python
pathway_fo = open('pathway.txt', 'r')

# format: {ko# : function}
ko_function_mapping = dict()

for row in pathway_fo:
    row_split = row.split('\t')
    ko_function_mapping['ko'+row_split[0][8:]] = row_split[1][:-1]    
```

```python
import operator
# exclude ko01100, ko01110 and ko01120 
excluded_pathway = ['ko01100', 'ko01110', 'ko01120']

# Start from pre clusters, there are 5 of them, 0 to 4
for i in range(5):
    # format: {ko# : count}
    pathway_count = dict()
    pre_cluster_fo = open('pre_cluster_ko/pre_c_{}.txt'.format(i),'r')
    
    # skip the header
    next(pre_cluster_fo)
    
    # row format: ['"ko00010"', '260922.863708514\n']
    for row in pre_cluster_fo:
        if row.split('\t')[0][1:-1] not in excluded_pathway:
            pathway_count[row.split('\t')[0][1:-1]] = float(row.split('\t')[1][:-1])
        
    # Find the top ten  
    top_ten = dict(sorted(pathway_count.items(), key=operator.itemgetter(1), reverse=True)[:10])
    
    # Generate the format that can be used in Google chart
    print('Pre-cluster {}:'.format(i))
    print("['Pathway', 'Counts'],")
    for ko, count in top_ten.items():
        print("['{}', {}],".format(ko_function_mapping[ko], count))
    
    pre_cluster_fo.close()
```
